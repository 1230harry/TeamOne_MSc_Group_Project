```python
#Customer Feedback chart
plt.hist(df['Rating'], bins=5)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Histogram of Ratings')
plt.show()
```

```python
# create scatter plot
plt.scatter(df['Uniq Id'], df['Rating'])
plt.xlabel('Uniq Id')
plt.ylabel('Rating')
plt.title('Scatter plot')
plt.show()
```

```python
# Lets check null for dataset
df.isnull().sum()
```

```python
# fill null values with mean
df.fillna(df.mean(), inplace=True)
```

```python
# Lets check null for dataset
df.isnull().sum()
```

```python
# create scatter plot
plt.scatter(df['Review Upvotes'], df['Review Downvotes'])
plt.xlabel('Review Downvotes')
plt.ylabel('Review Upvotes')
plt.title('Scatter plot')
plt.show()
```

```python
# check for null values after filling
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'Rating', y= 'Review Upvotes', data = df)
```
